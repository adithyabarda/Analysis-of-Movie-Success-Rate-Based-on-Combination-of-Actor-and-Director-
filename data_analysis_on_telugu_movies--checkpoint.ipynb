```python
df['Hero'] = df['Hero'].apply(lambda x: x.split(',')[0] if pd.notnull(x) else None)
df['Heroine'] = df['Heroine'].apply(lambda x: x.split(',')[0] if pd.notnull(x) else None)

hero_counts = df['Hero'].value_counts()
heroine_counts = df['Heroine'].value_counts()

print("Hero count:")
print(hero_counts)

print("\nHeroine count:")
print(heroine_counts)
```

# Combination Analysis of Hero and Heroine

```python
df['Hero_Heroine'] = df['Hero'] + ', ' + df['Heroine']


hero_heroine_counts = df['Hero_Heroine'].value_counts()

print("Combinations of Hero and Heroine:")
print(hero_heroine_counts)
```

```python
plt.figure(figsize=(10, 6))
hero_heroine_counts.plot(kind='bar', color='lightgreen')
plt.title('Hero-Heroine Combination Counts')
plt.xlabel('Hero-Heroine Combination')
plt.ylabel('Count')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(base=20))  
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
action_movies = df[df['Genre'].fillna('').str.contains('Action')]


director_action_counts = action_movies.groupby('Director').size()


sorted_director_action_counts = director_action_counts.sort_values(ascending=False)

print("Director                  Action Movies")
for director, action_count in sorted_director_action_counts.items():
    print(f"{director:25} {action_count}")
```

```python
df_cleaned = df.dropna(subset=['Runtime'])

df_cleaned['Runtime'] = pd.to_numeric(df_cleaned['Runtime'], errors='coerce')


mean_runtime = df_cleaned['Runtime'].mean()
median_runtime = df_cleaned['Runtime'].median()
mode_runtime = df_cleaned['Runtime'].mode()[0]  

print(f"Mean Runtime: {mean_runtime:.2f} minutes")
print(f"Median Runtime: {median_runtime:.2f} minutes")
print(f"Mode Runtime: {mode_runtime:.2f} minutes")
```

```python
values = [mean_runtime, median_runtime, mode_runtime]
labels = ['Mean', 'Median', 'Mode']


plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=['skyblue', 'lightgreen', 'lightcoral'])
plt.xlabel('Statistical Measures')
plt.ylabel('Runtime (minutes)')
plt.title('Statistical Measures of Movie Runtimes')
plt.ylim(0, max(values) * 1.2)  
plt.show()
```
